```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
agg_data = list()
data_path = "../data/crypto-data.csv"
crypto = pd.read_csv(data_path)
crypto_ = crypto[["tradeDate", "symbol", "avg_latest"]]
crypto_ = crypto_.pivot_table(index="tradeDate", columns="symbol", values = "avg_latest")
crypto2_ = crypto_.pct_change()
crypto2_ = crypto2_[3:] # remove na
```

```python
pipe = make_pipeline(StandardScaler(with_std=False), PCA())
pca = pipe.fit(crypto2_)  # apply scaling on training data
```

```python
scores = pipe["pca"].explained_variance_ratio_
s10 = scores[:10]
```

```python
df = pd.DataFrame({"scores": scores, "cumsum": scores.cumsum()})
# df["scores"].plot(title="crypto scores")
# df["cumsum"].plot(title="crypto scores cumulative")
```
